```python
def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()

def similar_jelly(a,b):
    return jellyfish.jaro_distance(a,b)
```

```python
def multipleReplace(text, wordDict):
    for key in wordDict:
        text = text.replace(key, wordDict[key])
    return text
```
